```python
import numpy as np
import matplotlib.pyplot as plt
```

**Mean Squared Error**

```python
def mean_squared_error(ytrue,ypred):
  cost = np.sum((ytrue-ypred)**2)/len(ytrue)
  return cost
```

**Gradient Descent**

```python
def gradientDescent(x,y,iter = 1000,learning_rate=0.0001,stopping_thresold=1e-6):
  cur_weight = 1
  bias = 0
  iterations = iter
  alpha = learning_rate
  n = float(len(x));
  costs = []
  weights = []
  b = []
  prev_cost = None
  for i in range(iterations):
    ypred = (cur_weight*x)+bias
    cur_cost = mean_squared_error(y,ypred)
    if(prev_cost and abs(prev_cost-cur_cost)<=stopping_thresold):
      break;
    prev_cost = cur_cost
    costs.append(cur_cost)
    weights.append(cur_weight);
    weight_grad = -(2/n)*sum(x*(y-ypred))
    bias_grad = -(2/n)*sum(y-ypred)
    cur_weight = cur_weight - (alpha*weight_grad);
    bias = bias - (alpha*bias)
    b.append(bias)
  plt.figure(figsize = (8,6))
  plt.plot(weights, costs)
  plt.scatter(weights, costs, marker='o', color='red')
  plt.title("Cost vs Weights")
  plt.ylabel("Cost")
  plt.xlabel("Weight")
  plt.show()
  return cur_weight, bias


```

```python
X = np.array([32.50234527, 53.42680403, 61.53035803, 47.47563963, 59.81320787,
           55.14218841, 52.21179669, 39.29956669, 48.10504169, 52.55001444,
           45.41973014, 54.35163488, 44.1640495 , 58.16847072, 56.72720806,
           48.95588857, 44.68719623, 60.29732685, 45.61864377, 38.81681754])
Y = np.array([31.70700585, 68.77759598, 62.5623823 , 71.54663223, 87.23092513,
           78.21151827, 79.64197305, 59.17148932, 75.3312423 , 71.30087989,
           55.16567715, 82.47884676, 62.00892325, 75.39287043, 81.43619216,
           60.72360244, 82.89250373, 97.37989686, 48.84715332, 56.87721319])
```

**Estimates and Visualization**

```python
estimated_weight,estimated_bias = gradientDescent(X,Y,iter=2000)
print(f"Estimated Weight: {estimated_weight}\nEstimated Bias: {estimated_bias}")
```
