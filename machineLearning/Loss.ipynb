```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Actual and Predicted values**

```python
actual_values= np.array([15, 25, 35, 45, 55])
predicted_values= np.array([18, 22, 38, 42, 52])
```

```python
number_of_samples = len(actual_values)
```

**L1 loss**

```python
l1 = np.sum(abs(actual_values-predicted_values))
print("L1 loss:",l1)
```

**Mean Absolute Error(MAE)**

```python
mae = l1/number_of_samples
print("Mean Absolute Error:",mae)
```

**L2 Loss**

```python
l2 = np.sum((actual_values-predicted_values)**2)
print("L2 loss:",l2)
```

**Mean Squared Error (MSE)**

```python
mse = l2/number_of_samples
print("Mean Square Error:",mse)
```

**Data visualization**

```python
coefficients = np.polyfit(actual_values, predicted_values, 1)
regression_line = np.poly1d(coefficients)

x_range = np.linspace(min(actual_values), max(actual_values), 100)
y_range = regression_line(x_range)

plt.figure(figsize=(8, 5))
plt.scatter(actual_values, predicted_values, color='blue', label='Data Points', s=100, alpha=0.8)

plt.plot(x_range, y_range, color='red', label=f'Regression Line (y={coefficients[0]:.2f}x+{coefficients[1]:.2f})')

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Scatter Plot with Regression Line')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
